Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data
Science (DS).

Answer. 1. Artificial Intelligence (AI):

Definition: AI is the broadest field that focuses on creating systems or machines that can perform tasks that usually require human intelligence.

Examples: Speech recognition, decision-making, playing chess, self-driving cars.

Goal: To simulate human thinking and behavior.


2. Machine Learning (ML):

Definition: ML is a subset of AI that focuses on enabling machines to learn from data and improve their performance over time without being explicitly programmed.

Examples: Email spam filters, recommendation systems, fraud detection.

Goal: To create algorithms that can find patterns in data and make predictions.


3. Deep Learning (DL):

Definition: DL is a subset of ML that uses neural networks with many layers (deep neural networks) to model complex patterns in data.

Examples: Image and speech recognition, language translation, autonomous vehicles.

Goal: To handle large-scale, unstructured data like images, audio, and text more effectively than traditional ML.


4. Data Science (DS):

Definition: DS is a multidisciplinary field that combines statistics, programming, domain knowledge, and data analysis to extract insights from data.

Examples: Business intelligence, customer analytics, trend forecasting.

Goal: To make data-driven decisions and solve real-world problems using data.

Question 2: What are the types of machine learning? Describe each with one
real-world example.

Answer. 1. Supervised Learning
Definition:

The model is trained on a labeled dataset, meaning the input data comes with the correct output. The goal is to learn a mapping from inputs to outputs.

 Real-World Example:

Email Spam Detection

Input: Email text (features)

Output: "Spam" or "Not Spam" (labels)

The model learns from many labeled emails and predicts whether new emails are spam.

2. Unsupervised Learning
 Definition:

The model is given unlabeled data and must find patterns, structures, or groupings within the data on its own.

 Real-World Example:

Customer Segmentation

Input: Customer data (age, income, purchase history)

No output labels provided

The algorithm groups customers into segments based on similar behavior (e.g., high spenders, occasional buyers).

3. Reinforcement Learning
 Definition:

An agent learns by interacting with an environment, receiving rewards or penalties based on its actions. The goal is to maximize cumulative reward over time.

 Real-World Example:

Self-Driving Cars

The car (agent) learns to drive by receiving feedback from the environment (e.g., rewards for staying in lane, penalties for collisions).

Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine
learning.

Answer. 1. Overfitting
Definition:

Overfitting occurs when a model learns the training data too well, including its noise and outliers, which causes it to perform poorly on new, unseen data.

 Characteristics:

High accuracy on training data

Low accuracy on testing data

Model is too complex (e.g., too many features or deep trees)

 Example:

A student memorizes practice questions instead of understanding concepts. In the exam (new data), they struggle with slightly different questions.

2. Underfitting
 Definition:

Underfitting occurs when a model is too simple to capture the underlying patterns in the data, leading to poor performance on both training and test data.

 Characteristics:

Low accuracy on both training and test data

Model is too simplistic (e.g., linear model for nonlinear data)

 Example:

A student only skims the textbook and lacks both detail and understanding. They perform poorly on both practice and exam questions.

3. Bias-Variance Tradeoff
 Definition:

The bias-variance tradeoff is a fundamental concept that describes the balance between two sources of error in a model:

Bias: Error due to overly simplistic assumptions in the model. Leads to underfitting.

Variance: Error due to model complexity and sensitivity to small changes in the training data. Leads to overfitting.

Question 4: What are outliers in a dataset, and list three common techniques for
handling them.

Answer. Outliers are data points that are significantly different from the rest of the data. They lie far outside the overall pattern of a distribution and can skew the results of data analysis or model training.

 Example: In a dataset of students' test scores ranging from 60–90, a score of 5 or 100 might be an outlier.  

 Three Common Techniques for Handling Outliers:
1. Removal (Deleting Outliers)

How: Identify outliers using statistical methods (like Z-score or IQR) and remove them.

When to Use: When outliers are due to errors or noise, not relevant to the problem.

Example: Removing a negative age from a customer dataset.

2. Transformation or Capping (e.g., Winsorization)

How: Replace extreme outlier values with a specified percentile value (e.g., 5th and 95th percentiles).

When to Use: When you want to reduce the influence of outliers without removing them.

Example: Capping incomes above the 99th percentile in financial data.

3. Imputation or Replacement

How: Replace outliers with the mean, median, or mode of the feature.

When to Use: When you suspect the outlier is a data entry error or missing value substitute.

Example: Replacing an outlier height of 300 cm with the average height of 170 cm.


Question 5: Explain the process of handling missing values and mention one
imputation technique for numerical and one for categorical data.

Answer. Missing values occur when no data is stored for a feature (column) in a dataset. These gaps can arise due to:

Data entry errors

Sensor failures

User non-responses in surveys

Handling them properly is critical because many machine learning algorithms do not work well with missing data.

 Process of Handling Missing Values:

Identify Missing Values

Use methods like .isnull() in Python (Pandas) to detect missing data.

Understand the Cause

Determine if missingness is random or due to a pattern (e.g., missing only for certain groups).

Choose a Handling Strategy:

Remove missing data (only if few and not significant).

Impute (fill in) missing data using statistical techniques.

 Imputation Techniques:
 1. Numerical Data Imputation:

Technique: Mean Imputation

How it works: Replace missing numerical values with the mean of the non-missing values in that column.

Example: If the "age" column has missing values, calculate the average age and fill the missing cells with that average.


 2. Categorical Data Imputation:

Technique: Mode Imputation

How it works: Replace missing categorical values with the most frequent category (mode).

Example: If the "gender" column has missing values, fill them with the most common value like "Male" or "Female".

```python
 #Question 8 : Write a Python script to:
#● Generate 1000 samples from a normal distribution.
#● Introduce 50 random missing values.
#● Fill missing values with the column mean.
#● Plot a histogram before and after imputation.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

data_with_nan = data.copy()
missing_indices = np.random.choice(1000, 50, replace=False)
data_with_nan[missing_indices] = np.nan


df = pd.DataFrame({'Value': data_with_nan})

#plot histogram
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df['Value'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Before Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Fill missing values with the column mean
mean_value = df['Value'].mean()
df['Value'].fillna(mean_value, inplace=True)

#  Plot histogram AFTER imputation
plt.subplot(1, 2, 2)
plt.hist(df['Value'], bins=30, color='lightgreen', edgecolor='black')
plt.title('After Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
